```python
plt.figure(figsize=(20,25),facecolor='green')
plotnumber=1
for column in df1:
  if plotnumber<=17:
    ax=plt.subplot(6,3,plotnumber)
    plt.scatter(df1[column],y)
    plt.xlabel(column,fontsize=20)
    plt.ylabel('W',fontsize=20)
  plotnumber+=1
plt.tight_layout()
```

```python
scalar=StandardScaler()
x_scaled=scalar.fit_transform(x)
x_scaled
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.35,random_state=42)
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
f='finalized_model.pickle'
pickle.dump(regression,open(f,'wb'))
loaded_model=pickle.load(open(f,'rb'))
a=loaded_model.predict(scalar.transform([[724,5575,1497,300,42,139,383,973,104,641,601,3.73,2,8,56,88]]))
a
```
